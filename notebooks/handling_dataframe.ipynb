# Import modules

````python
import numpy as np
import pandas as pd
````

# Generate data

````python
df = pd.DataFrame(np.random.randn(20, 5), columns=list('ABCDE'))
df
````

# 列(行) 操作

apply

```
Args:
    axis: 0: 列, 1: 行 に適用する

Return:
    (pandas.Series)
```

````python
df.apply(lambda x: sum(x), axis=1)
````

# 列追加

assign

```
Returns:
    (pandas.DataFrame)
```

DataFrame に対して、操作関数を適用するイメージ

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)  # A-E の和
)
````

# 列選択

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]]
````

# 列名変更

rename

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
})
````

# 集計関数

groupby, max

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max()
````

# Index の振り直し

reset_index

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index()
````

"id" を Index とする

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id")
````

# select where

pipe

Where 句によるデータ絞り込みのイメージ

絞り込み以外の(SQL の join のような)処理も書ける

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id").pipe(
    lambda x: x[x.value > -2]
)
````

df.pipe().pipe() のように条件を追加していくことが可能

以下では、絞り込み条件というよりは、変換処理のような感じで join で結合している

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id").pipe(
    lambda x: x[x.value > -2]
).pipe(
    lambda x: x.add_suffix("_l").join(x.add_suffix("_r"))
)
````

# 縦持ち から 横持ち へ変換

pivot

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id").pipe(
    lambda x: x[x.value > -2]
).pipe(
    lambda x: x.add_suffix("_l").join(x.add_suffix("_r"))
).reset_index().groupby(["id", "key_l"]).sum().reset_index().assign(
    key=lambda x: x.value_l.map(lambda x: x * 10).round().map(lambda x: "label" + str(int(x)))
).rename(columns={
    "value_r": "value"
})[["key", "value"]].reset_index()
````

↑を縦横変換

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id").pipe(
    lambda x: x[x.value > -2]
).pipe(
    lambda x: x.add_suffix("_l").join(x.add_suffix("_r"))
).reset_index().groupby(["id", "key_l"]).sum().reset_index().assign(
    key=lambda x: x.value_l.map(lambda x: x * 10).round().map(lambda x: "label" + str(int(x)))
).rename(columns={
    "value_r": "value"
})[["key", "value"]].reset_index().pivot(
    "index",
    "key",
    "value"
)
````

# 横持ち から 縦持ち へ変換

stack

```
Return:
    (pandas.Series) 階層インデックスを持つ pandas.Series
```

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id").pipe(
    lambda x: x[x.value > -2]
).pipe(
    lambda x: x.add_suffix("_l").join(x.add_suffix("_r"))
).reset_index().groupby(["id", "key_l"]).sum().reset_index().assign(
    key=lambda x: x.value_l.map(lambda x: x * 10).round().map(lambda x: "label" + str(int(x)))
).rename(columns={
    "value_r": "value"
})[["key", "value"]].reset_index().pivot(
    "index",
    "key",
    "value"
).fillna(0).astype(np.int8).stack()
````

pipe によって Series を DataFrame にする

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id").pipe(
    lambda x: x[x.value > -2]
).pipe(
    lambda x: x.add_suffix("_l").join(x.add_suffix("_r"))
).reset_index().groupby(["id", "key_l"]).sum().reset_index().assign(
    key=lambda x: x.value_l.map(lambda x: x * 10).round().map(lambda x: "label" + str(int(x)))
).rename(columns={
    "value_r": "value"
})[["key", "value"]].reset_index().pivot(
    "index",
    "key",
    "value"
).fillna(0).astype(np.int8).pipe(
    lambda x: pd.DataFrame({"value": x.stack()})
)
````

value が 0 以外のもののみ返す

````python
df.assign(
    round_A=lambda x: x.A.round(),
    round_B=lambda x: x.B.round(),
    total=lambda x: x.apply(lambda row: sum(row), axis=1)
)[["round_A", "round_B", "total"]].rename(columns={
    "round_A": "id",
    "round_B": "key",
    "total": "value"
}).groupby(["id", "key"]).max().reset_index().set_index("id").pipe(
    lambda x: x[x.value > -2]
).pipe(
    lambda x: x.add_suffix("_l").join(x.add_suffix("_r"))
).reset_index().groupby(["id", "key_l"]).sum().reset_index().assign(
    key=lambda x: x.value_l.map(lambda x: x * 10).round().map(lambda x: "label" + str(int(x)))
).rename(columns={
    "value_r": "value"
})[["key", "value"]].reset_index().pivot(
    "index",
    "key",
    "value"
).fillna(0).astype(np.int8).pipe(
    lambda x: pd.DataFrame({"value": x.stack()})
).pipe(
    lambda x: x[x.value != 0]
)
````
